```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Load the dataset
file_path = 'C:\\Users\\anoos\\Downloads\\All the Intermediate Colleges in Pakistan.csv'
df = pd.read_csv(file_path)

# Clean the 'Rating' column
df['Rating'] = df['Rating'].str.replace('*', '')  # Remove asterisks
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')  # Convert to numeric

# Handle missing values in 'Location' and 'Sector'
df['Location'] = df['Location'].fillna('Unknown')
df['Sector'] = df['Sector'].fillna('Unknown')

# Drop rows with missing 'Rating' values for analysis
df = df.dropna(subset=['Rating'])
```

# Descriptive statistics

```python

# Standard Deviation
print("Standard Deviation of Ratings: ", np.std(df['Rating']))
```

# Statistical Tests
Shapiro-Wilk Test

```python
# Shapiro-Wilk Test
shapiro_test = stats.shapiro(df['Rating'])
print("Shapiro-Wilk Test:", shapiro_test)
```

K2 Test (D'Agostino's K-squared test)

```python
# K2 Test
k2_test = stats.normaltest(df['Rating'])
print("K2 Test:", k2_test)
```

Kolmogorov-Smirnov Test

```python
# Kolmogorov-Smirnov Test
ks_test = stats.kstest(df['Rating'], 'norm', args=(np.mean(df['Rating']), np.std(df['Rating'])))
print("Kolmogorov-Smirnov Test:", ks_test)
```

ANOVA Test

```python
# ANOVA
anova_results = ols('Rating ~ Sector', data=df).fit()
anova_table = sm.stats.anova_lm(anova_results, typ=2)
print("ANOVA Table:")
print(anova_table)
```

Tukey's HSD Test

```python
# Tukey's HSD Test
tukey = pairwise_tukeyhsd(endog=df['Rating'], groups=df['Sector'], alpha=0.05)
print(tukey)
```

Chi-Squared Test

```python
# Chi-Squared Test
contingency_table = pd.crosstab(df['Sector'], df['Rating'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print("Chi-Squared Test:")
print(f"Chi2: {chi2}, p-value: {p}, Degrees of Freedom: {dof}")
```

T-tests

```python
# T-tests (between two sectors)
public_ratings = df[df['Sector'] == 'Public']['Rating']
private_ratings = df[df['Sector'] == 'Private']['Rating']
t_test_result = stats.ttest_ind(public_ratings, private_ratings)
print("T-test between Public and Private Sector Ratings:", t_test_result)
```

# Outlier Analysis

```python
# Outlier Analysis
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Rating'])
plt.title('Outlier Analysis of Ratings')
plt.xlabel('Rating')
plt.show()

# Identifying Outliers
q1 = df['Rating'].quantile(0.25)
q3 = df['Rating'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = df[(df['Rating'] < lower_bound) | (df['Rating'] > upper_bound)]
print("Outliers:")
print(outliers)
```
